```python
import pandas as pd
from datetime import datetime

# Read the Excel file into a DataFrame
df = pd.read_excel('Data_GamesCompanies/Data_ALL.xlsx')
```

```python
df.info()
```

```python
# Columns to convert to numeric
numeric_columns = ['Revenue/Downloads', 'Usage Penetration', 'Avg Sessions / User']

# Convert specified columns to numeric
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Columns to convert from date to amount number
date_columns = ['Avg Session Duration', 'Avg Time / User']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
    df[col] = df[col].apply(lambda x: (x.hour * 60) + (x.minute) + (x.second / 60) if not pd.isnull(x) else None)

# Example time count conversion
time_count = '1900-01-02 00:04:10.992000'
time_obj = datetime.strptime(time_count, '%Y-%m-%d %H:%M:%S.%f')
total_minutes = time_obj.second // 60 + time_obj.minute + time_obj.hour * 60
print(total_minutes)

# Now you can work with the transformed DataFrame 'df'
df.head(3)
```

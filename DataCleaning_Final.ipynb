# Initialisation + main VARIABLE

# Armonisation of column names

```python
# Defining Exact Names for each columns
for date in date_id:
    df = pd.read_excel(f'Data_GamesCompanies/Data_{date}.xlsx')
    columns_names = [
        '#', 'Unified App', 'Classification', 'Downloads', 
        'Cumulative Downloads', 'Revenue', 'Cumulative Revenue', 
        'Revenue/Downloads', 'Paid Downloads', 'Organic Downloads', 
        '% Paid Downloads', '% Organic Downloads', 'Usage Penetration', 
        'Active Users', 'Install Penetration', 'Install Base', 
        'Open Rate', 'Avg Sessions / User', 'Avg Session Duration', 
        'Avg Time / User', 'Total Time', 
        'Avg Active Days', '% Active Days', 'Genre', 
        'Company', 'Year', 'Month'
    ]
    df.columns = columns_names

    # Save the resulting DataFrame to a new Excel file
    df.to_excel(f'Data_{date}.xlsx', index=False)

df.head(3)
```

# Removing unwanted characters

```python
for date in date_id:
    df = pd.read_excel(f'Data_{date}.xlsx')

    # Specify the columns to clean
    columns_to_clean = ['Downloads', 'Cumulative Downloads', 'Revenue', 'Cumulative Revenue', 'Revenue/Downloads'
                        'Paid Downloads', 'Organic Downloads', 'Active Users', 'Install Base', 'Total Time']
    # Remove unwanted characters and convert numbers to their numerical equivalent
    for column in columns_to_clean:
        try:
            df['Revenue/Downloads'] = df['Revenue/Downloads'].str.replace(r'[\,]', '.', regex=True)  # Remove ',' with '.'
        except:
            pass
        try:
            df[column] = df[column].str.replace(r'[\$]', '', regex=True)  # Remove '$'
        except:
            pass
        try:
            df[column] = df[column].str.replace(r'(\d+)([kK])', r'\g<1>e3', regex=True)  # Convert 'k' to '*1e3'
        except:
            pass
        try:
            df[column] = df[column].str.replace(r'(\d+)([mM])', r'\g<1>e6', regex=True)  # Convert 'm' to '*1e6'
        except:
            pass
        try:
            df[column] = df[column].str.replace(r'(\d+)([bB])', r'\g<1>e9', regex=True)  # Convert 'b' to '*1e9'
        except:
            pass
        try:
            df[column] = pd.to_numeric(df[column])  # Convert the column to numeric data type
        except:
            pass

    # Save the resulting DataFrame to a new Excel file
    df.to_excel(f'Data_{date}.xlsx', index=False)

df.head(3)
```

# Datetime to amount in minuts

```python
from datetime import datetime

# Columns to convert from date to amount number
date_columns = ['Avg Session Duration', 'Avg Time / User']

for date in date_id:
    df = pd.read_excel(f'Data_{date}.xlsx')

    for col in date_columns:
        df[col] = pd.to_datetime(df[col], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
        df[col] = df[col].apply(lambda x: (x.hour * 60) + (x.minute) + (x.second / 60) if not pd.isnull(x) else None)

    # Save the resulting DataFrame to a new Excel file
    df.to_excel(f'Data_{date}.xlsx', index=False)

df.head(3)
```

```python
# Columns to convert to numeric
numeric_columns = ['Revenue/Downloads', 'Usage Penetration', 'Avg Sessions / User']

for date in date_id:
    df = pd.read_excel(f'Data_{date}.xlsx')

    # Convert specified columns to numeric
    for col in numeric_columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Save the resulting DataFrame to a new Excel file
    df.to_excel(f'Data_{date}.xlsx', index=False)

df.info()
```

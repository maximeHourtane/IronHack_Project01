```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Data_GamesCompanies/Data_ALL.xlsx')

correlation_matrix = df.corr()  # Calculate the correlation matrix

fig, ax = plt.subplots(figsize=(10, 10))  # Adjust figsize to accommodate all columns
heatmap = ax.matshow(correlation_matrix, cmap='coolwarm')  # Generate the heatmap

# Customize the plot
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.colorbar(heatmap)  # Add a colorbar

plt.show()  # Show the plot
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Excel file into a DataFrame
df = pd.read_excel('Data_GamesCompanies/Data_ALL.xlsx')

# Select only the numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Create a box plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numeric_columns])
plt.title("Box Plot of Numeric Columns")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Excel file into a DataFrame
df = pd.read_excel('Data_GamesCompanies/Data_ALL.xlsx')

# Create a box plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df['Downloads'])
plt.title("Box Plot of Downloads")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by Genre
grouped = df.groupby('Genre')

# Create separate DataFrames for each genre
genre_dataframes = {}
for genre, group in grouped:
    genre_dataframes[genre] = group.copy()

    correlation_matrix = df.corr()  # Calculate the correlation matrix

    fig, ax = plt.subplots(figsize=(10, 10))  # Adjust figsize to accommodate all columns
    heatmap = ax.matshow(correlation_matrix, cmap='coolwarm')  # Generate the heatmap

    # Customize the plot
    plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
    plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
    plt.colorbar(heatmap)  # Add a colorbar
    print(f"Genre: {genre}")

    plt.show()  # Show the plot

# # Print the separate DataFrames
# for genre, genre_df in genre_dataframes.items():
#     print(f"Genre: {genre}")
#     print(genre_df)
#     print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Data_GamesCompanies/Data_ALL.xlsx')

correlation_matrix = df.corr()  # Calculate the correlation matrix

fig, ax = plt.subplots(figsize=(10, 10))  # Adjust figsize to accommodate all columns
heatmap = ax.matshow(correlation_matrix, cmap='coolwarm')  # Generate the heatmap

# Customize the plot
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.colorbar(heatmap)  # Add a colorbar

plt.show()  # Show the plot
```

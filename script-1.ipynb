```python
def get_game_company():
    fp = open("games_company.txt", "r")
    game = []
    company = []
    for line in fp.readlines():
        l_array = line.split('|')
        game.append(l_array[0])
        company.append(l_array[1])
    return game, company

game, company = get_game_company()
for i in range(len(game)):
    print(f"{game[i]} | {company[i]}")
```

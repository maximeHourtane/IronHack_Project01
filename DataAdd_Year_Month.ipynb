```python
import pandas as pd

# List of DataSources
date_id = ["2307", "2306", "2305", "2304", "2303", "2302", "2301", "2212", "2211", "2210", "2209", "2208"]

for date in date_id:
    # 1 - Importing all Cleaned Data from excel sheets
    df = pd.read_excel(f'Data_GamesCompanies/Data_{date}.xlsx')

    # Select the 2 first digits of the 'date_id'
    year_digits = date[:2]
    # Add a 'Year' column by concatenating '20' with 'year_digits'
    df['Year'] = '20' + year_digits

    # Select the 2 last digits of the 'date_id'
    month_digits = date[-2:]
    # Add a 'Month' column using 'month_digits'
    df['Month'] = month_digits

    # Save the DataFrame
    df.to_excel(f'Data_GamesCompanies/Data_{date}.xlsx', index=False)

df
```
